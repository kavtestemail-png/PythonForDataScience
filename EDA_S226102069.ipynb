```python
# Part F
plt.hist(df["avg_match_quality"].dropna(), bins=20)
plt.show()
from scipy.stats import shapiro
shapiro(df["avg_match_quality"])
df["avg_match_quality"].skew()

df.boxplot("avg_match_quality")
```

# Part F
This seems to be a left/negatively skewed distribution as the tail tends to be on the left side. And the Skew value result of avg_match_quality(-0.2863029544181897) variable confirms it.
 and the shapiro wilk test result (P-value <0.05) confirms that the data are not normal frp avg_match_quality variable. 

 And regarding outliers it seems to be lot of outliers on 0 - 0.4 range as it can be seen in the plotted boxplot

```python
# Part G

plt.scatter(df["avg_match_quality"], df["employee_count"])
plt.show()
df["avg_match_quality"].corr(df["employee_count"])


from scipy.stats import pearsonr
corr, p_value = pearsonr(df["avg_match_quality"], df["employee_count"])
print("Correlation:", corr, ", P-value:", p_value)
```

As per the above plot and and the correlation co-efficient value and the P-value (<0.05) from pearson test it can be stated that these two numerical variables (avg_match_quality, employee_count) are not correlated under 95% significance
