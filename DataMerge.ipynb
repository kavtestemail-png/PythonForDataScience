```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv(r"response_data.csv")
df2 = pd.read_csv(r"school_data.csv")

df1.head()
```

```python
df2.head()
```

```python
# Merging the df1 and df2 to merge dataframe

df = pd.merge(df1,df2, on= "School ID")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum() 
# This means there's no missing values in our merged dataset
```

```python
# Dropping SchoolID since School is already there in data
df = df.drop(['School ID','Board ID'], axis=1)
df.head()
```

```python
# Since missing values aren't occurring naturally I will be removing 0-20% values in selected columns randomly as below


columns_to_modify = ["Board", "Level 1 (%)", "Q3(%)"]


# for col in columns_to_missing:
#     mask = np.random.rand(len(df)) < 0.2   # randomly selecting 20% of the length
#     df.loc[mask, col] = np.nan

for col in columns_to_modify:
    p = np.random.uniform(0.1,0.2)
    mask = np.random.rand(len(df)) < p
    df.loc[mask, col] = np.nan

print(df.head())
```

```python
# df.isnull().sum() 
(df.isnull().sum() / len(df)) * 100
```

```python
# Missing values Analysis (Q3(%), Level 1 (%), Board)

# Count Plot for Board
sns.countplot(x='Board', data=df)
plt.show()
# From here we can see Mode is Toronto DSB 
```

```python
df['Board'].mode()[0] # Most occurred Board, this will be used to replace missing values in board column
df['Board'] = df['Board'].fillna(df['Board'].mode()[0])
```

```python
df['Board'].describe()
```

```python
sns.countplot(x='Board', data=df)
plt.show()
```

```python
# Plotting distribution of Q3(%) field
df['Q3(%)'].hist()
plt.show()

# help(plt.hist)
```

```python
print("Mean of Q3(%) : ", df['Q3(%)'].mean())
print("Median of Q3(%) : ", df['Q3(%)'].median())
print("Mode of Q3(%) : ", df['Q3(%)'].mode()[0])

# Based on the histogram we can assume distribution is almost normal and Mean, Median and Mode are also very close values
```

```python
# Since all values are almost same Median was taken to replace missing values as this look like a slight positively skewed distribution

df['Q3(%)'] = df['Q3(%)'].fillna(df['Q3(%)'].median())
```

```python
# Plotting distribution of missing filled Q3(%) field
df['Q3(%)'].hist()
plt.show()
```

```python
# Plotting distribution of Level 1 (%) field
df['Level 1 (%)'].hist()
plt.show()
```

```python
print("Mean of Level 1 (%) : ", df['Level 1 (%)'].mean())
print("Median of Level 1 (%) : ", df['Level 1 (%)'].median())
print("Mode of Level 1 (%) : ", df['Level 1 (%)'].mode()[0])

# Based on the distribution since its positively skewed distribution rather that using mean using median is better for missing value filling
```

```python
df['Level 1 (%)'] = df['Level 1 (%)'].fillna(df['Level 1 (%)'].median())
```

```python
# Plotting distribution of of missing filled Level 1 (%) field
df['Level 1 (%)'].hist()
plt.show()
```
